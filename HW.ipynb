Вводятся числа по одному на строке, пока не встретится пустая строка. Вывести количество совершенных чисел.

```python
def is_perfect(a):
    sum_divs = 0
    for i in range(1, a):
        if a % i == 0:
            sum_divs+=i
    if sum_divs == a:
        return True
    return False
        
a = input()
arr_num = []
while a != "":
    arr_num.append(int(a))
    a = input()
for num in arr_num:
    if is_perfect(num):
        print(num)
```
